```python
def P(x):
    x0=-1.0
    x1=-0.4
    x2=0.2
    x3=0.8
    y0=0.606531
    y1=0.923116
    y2=0.980199
    y3=0.726149
    
    L1=y3*((x-x0)*(x-x1)*(x-x2))/((x3-x0)*(x3-x1)*(x3-x2))
    L2=y2*((x-x0)*(x-x1)*(x-x3))/((x2-x0)*(x2-x1)*(x2-x3))
    L3=y1*((x-x0)*(x-x2)*(x-x3))/((x1-x0)*(x1-x2)*(x1-x3))
    L4=y0*((x-x1)*(x-x2)*(x-x3))/((x0-x1)*(x0-x2)*(x0-x3))
    
    return L1+L2+L3+L4

print('f(0)=' , P(0))

x=np.linspace(-1,1,50) #δημιουργεί πίνακα 50 στοιχείων ισομοιρασμένα στο [-1,1]
plt.plot(x,P(x))
```
